Problem 1: Mobile Recharge Validation System


```python
def validate_recharge():
    valid_plans = [199, 299, 399, 599] # list of Plaans
    while True:
        amount = int(input("Enter recharge amount: "))# Take input of user plan to recharge
        if amount < 50:  # checks the plan is less 50 or not
            print("Amount must be at least ₹50. Try again.")
            continue
        if amount not in valid_plans: # check plans available in list of plans
            print("Invalid plan. Available plans:", valid_plans)
            continue
        print("Recharge successful!") # print recharge successful
        break

validate_recharge()
```

Problem 2: Inventory Reorder Alert System


```python
def check_inventory(products):
    for item, stock in products.items(): # iterate over the dictionary of products using items and stocks available
        if stock < 15: #checks if stocks less than the 15 reorder items alert comes
            print(item, "→ Reorder Alert")
        else:                             #otherwise stocks OK
            print(item, "→ Stock OK")

products = {"Shoes":10, "Shirts":25, "Watches":12 ,"Pants":16}
check_inventory(products)
```

 Problem 3: Student Result Processing System


```python
def check_result(marks):
    total = 0
    for m in marks:  # calculate the sum of the marks or using built in fuction sum()
        total += m
    avg = total / len(marks) # calculate the average marks
    if avg >= 50: # checks if avg greater than 50 return Pass
        return "Pass"
    else:         # otherwise return Fail
        return "Fail"

print("Final Result : ",check_result([60,70,50,40]))
```

Problem 4: Cab Fare Estimator with Retry Option


```python
def calculate_fare():
    while True:
        km = float(input("Enter distance in km: "))  # get the distance from the user and convert to a float for decimal support
        peak = input("Peak hour? (yes/no): ").lower() # check if the peak hour pricing applies
        fare = 50 + (12 * km) # calculate the fare price
        if peak == "yes":  # apply a 25% surcharge if it's peak hour
            fare *= 1.25
        print("Total Fare:", round(fare,2))  # final amount rounded to 2 decimal
        retry = input("Calculate again? (yes/no): ").lower()  # ask the user if they want to run another calculation
        if retry != "yes":
          print("Thank you for using the fare calculator!")
          break

calculate_fare()
```

Problem 5: Employee Attendance Eligibility Checker

```python
def attendance_check(attendance):
    present = 0
    for day in attendance: # iterate through the list and count occurrences of "P" (Present)
        if day == "P":
            present += 1
    percentage = (present / len(attendance)) * 100 #calculate the percentage
    if percentage >= 75: # check if percentage is greater than 75% for Eligible otherwise Not
        return "Eligible"
    else:
        return "Not Eligible"

print("Employee Attendance Eligibility:",attendance_check(["P","P","A","P","P"]))
```

Problem 6: Password Strength Checker

```python
def check_password(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"

    if len(password) < 8: # check if the password meets the minimum length requirement
        return "Weak Password"

    for char in password: # scan each character in the password to check for required types
        if char.isdigit(): # check if password has the numeric digits
            has_digit = True
        if char in special_chars: # check if password has tghe special characters
            has_special = True

    if has_digit and has_special: # if the password has the both special characters and numerics decides as Strong
        return "Strong Password"
    else:                         # otherwise Weak
        return "Weak Password"

password = input("Enter password: ")
print(check_password(password))
```
