Problem Statement 1: Employee Performance Bonus Eligibility


```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

highest_score = max(employees.values()) # It stores the highest performance score

top_performers = [] #List to store the top performers
for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)

print("Top Performers Eligible for Bonus:", ", ".join(top_performers), f"(Score: {highest_score})")
```

Problem Statement 2: Search Query Keyword Analysis

```python
query = "Buy mobile phone buy phone online" # input Query
query = query.lower() # tzhis converts query to lowercase letters
import string # import string
query = query.translate(str.maketrans('', '', string.punctuation)) # remove the extra characters and punctuations

words = query.split() # split the sentences into words

frequency = {} # dictionary to store the word and frequency of the word
for word in words: # count frequency
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1

result = {}
for word, count in frequency.items(): #for more than once in sentence
    if count > 1:
        result[word] = count
print(result)
```

Problem Statement 3: Sensor Data Validation

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
valid_readings = [] # List to store the valid_readings
for index, value in enumerate(sensor_readings):
    if value % 2 == 0:   # check even number
        valid_readings.append((index, value))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

Problem Statement 4: Email Domain Usage Analysis

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]
domain_count = {}
for email in emails:
    domain = email.split("@")[1]   # extract domain by split the gmail by delimiter '@' and access the domain name by indexing 1
    if domain in domain_count:    # frequency count for the domain
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

total_users = len(emails) # To calculate the percentage

for domain, count in domain_count.items():
    percentage = (count / total_users) * 100
    print(f"{domain}: {int(percentage)}%")
```

Problem Statement 5: Sales Spike Detection

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

average_sales = sum(sales) / len(sales) #calculate the average

for day, value in enumerate(sales, start=1): #enumerate the sales and compare the value greater than the avg
    if value > average_sales:
        print(f"Day {day}: {value}")
```

Problem Statement 6: Duplicate User ID Detection

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
frequency = {} # dictionary to store the frequency

for user in user_ids:
    if user in frequency:
        frequency[user] += 1
    else:
        frequency[user] = 1

for user, count in frequency.items(): # to display the duplicates
    if count > 1:
        print(f"{user} → {count} times")
```
