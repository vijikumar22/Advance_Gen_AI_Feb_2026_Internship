Problem 1: Social Media – Post Engagement Analyzer


```python
def post_engagement(likes): # fuction to calculate the likes
    total_likes = 0
    for like in likes: # For calculate the total likes
        total_likes += like

    print("Total Likes:", total_likes)

    if total_likes >= 1000: # checks if posts has higher than the 1000 prints Vral Post
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement") # otherwise print Normal Engagement

post_engagement([200,300,250,300])  # Pass the list of likes to function
```

Problem 3: Agriculture – Rainfall Adequacy Checker


```python
def rainfall_check(rainfall_data, required): # function to check the seasonal rainfall data
    avg = sum(rainfall_data)/len(rainfall_data) # calculate the avg of rainfall
    print("Average Rainfall:", int(avg)) # print the avg

    if avg >= required: # checks if avg is greater than required print Adequate
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall") #otherwise Inadequate

rainfall_check([70,75,80,65],70) # pass the list of rainfall data and required
```

Problem 4: Social Media – Duplicate Account Detection


```python
def duplicate_accounts(users):
    # checks the length of the original list and  length of set(users)  set contains only uniques values
    if len(users) != len(set(users)):
        print("Duplicate Accounts Found: Yes") # if not equal it has the Duplicates Accounts
    else:
        print("Duplicate Accounts Found: No") # otherwise No

duplicate_accounts(["user1","user2","user1"])
```

Problem 6: Agriculture – Premium Crop Price Filter

```python
def premium_crops(prices): # function to filters crop prices above a premium threshold
    premium = []  # list to add the premium crops
    for price in prices:
        if price > 2000: # if price > 2000 is considered as premium add to the new list
            premium.append(price)
    print("Premium Crops:", premium) # print the new premium

price=[1500,2500,3200,1800]
premium_crops(price)
```

Problem 7: System Monitoring – Application Health

```python
def app_health(errors): # function to evaluate application health based on error count
    print("Error Count:", errors)

    if errors == 0:  # Check if the system is perfectly clean
        status = "Healthy"
    elif errors <= 5: # Check if errors are within a manageable threshold (1 to 5)
        status = "Minor Issues"
    else:              # Fallback for any error count higher than 5
        status = "Critical Issues"

    print("System Status:", status)

app_health(7)
```

Problem 9: E-Learning – Student Attendance Eligibility System


```python
def exam_eligibility(attendance):
    present = attendance.count("P") # count the no of present "P"
    percentage = (present/len(attendance))*100 # calculate the percentage

    print("Attendance Percentage:", percentage)

    if percentage >= 75:  # check if percentage is greater than 75% print Eligible
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible") # otherwise Not Eligible

exam_eligibility(["P","P","A","P","P"])
```

Problem 10: Smart Electricity Bill Analyzer


```python
def electricity_bill(units):
    bill = 0

    if units <= 100:
        bill = units * 3  # first 100 units at $3 each
    elif units <= 200:
        bill = (100*3) + (units-100)*5  # first 100 units at $3, remaining units at $5
    else:
        bill = (100*3) + (100*5) + (units-200)*7  # first 100 at $3, next 100 at $5, anything over 200 at $7

    # Categorize the total cost into usage levels
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    print("Electricity Bill:", bill)
    print("Usage Level:", usage)

electricity_bill(250)
```
